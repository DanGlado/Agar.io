#### Напишите программу, которая отображает квадратное окно размером 600 на 600, полностью заполненное областью для рисования. По центру области должна отображаться окружность радиусом 200. Вдоль этой окружности по часовой стрелке должна двигаться точка.

```python
from tkinter import *
import math
root = Tk()
c = Canvas(root, width=600, height=600, bg="white")
c.pack()
main_ball = c.create_oval(200, 200, 400, 400, fill='red')
small = c.create_oval(275, 175, 325, 225, fill='green')
i = 1
def motion():
    global i
    angle = i * math.pi / 180
    x = 5 * math.cos(angle)
    y = 5 * math.sin(angle)
    c.move(small, x, y)
    if True:
        root.after(50, motion)
        i+=3

motion()
root.mainloop()
```

#### Попробуйте изменить скорость и направление движения точки. Сделайте переменную, которая задает скорость и направление движения.


```python
from tkinter import *
import math
root = Tk()
c = Canvas(root, width=600, height=600, bg="white")
c.pack()
r = 5

main_ball = c.create_oval(200, 200, 400, 400, fill='red')
small = c.create_oval(275, 175, 325, 225, fill='green')
i = 1
ms = 100

def reverse():
    global small
    global r
    global i
    r *= -1
    i = 1
    c.delete(small)
    small = c.create_oval(275, 175, 325, 225, fill='green')
button1 = Button(root, bg="red", text=u"Изменить направление!", command=reverse)
button1.pack()

def speed():
    global ms
    if ms > 15:
        ms -= 15

button2 = Button(root, bg="red", text=u"Увеличить скорость!", command=speed)
button2.pack()

def motion():
    global i
    angle = i * math.pi / 180

    x = r * math.cos(angle)  # против часовой -5
    y = abs(r) * math.sin(angle)
    c.move(small, x, y)
    print(x, y, i)
    if True:
        root.after(ms, motion)  # Скорость вращения
        i += 3

motion()
root.mainloop()
```
